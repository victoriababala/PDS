```python
#5
import numpy as np
x = np.arange(-10,10,0.1)

y1 = [i**2+200 for i in x]
y2 = [abs(i)*100 for i in x]
```

```python
#6
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
#7
plt.plot(x,y1, ls='dotted')
plt.plot(x,y2, ls = '--')
plt.show()
```

```python
#8
plt.plot(x,y1, color ='red')
plt.plot(x,y2, c = '#522abf')
plt.show()
```

```python
#9
plt.plot(x,y1, lw = 3)
plt.plot(x,y2, lw = 5.5)
plt.show()
```

```python
import math
```

```python
#10
y3 = [100 * math.sin(i) + 150 for i in x]
plt.fill_between(x,y3)
plt.show()
```
