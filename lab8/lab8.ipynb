```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('lab5.csv', sep=",")
df.head()
```

```python
#1
group_by_month = df.groupby('Month').size()
group_by_month.head()
group_by_month.plot(kind='bar')
plt.xlabel('Місяць')
plt.ylabel('Кількість жертв')
plt.title('Кількість жертв по місяцях')
plt.show()
# max victims num in 6-th month
```

```python
#1
sns.countplot(x='Month', data =df)
```

```python
sns.countplot(x='Month', data =df,  hue= "Vict Sex", palette=["firebrick", "navy"] )
```

```python
sns.catplot(x='Month', data =df, kind ='count')
```

```python
sns.catplot(x='Month', data =df, kind ='count', hue_order=['M','F'])
```

```python
# 2
monthly_avg_age = df.groupby('Month')['Vict Age'].mean()
monthly_avg_age.plot(kind='bar')
plt.xlabel('Місяць')
plt.ylabel('Середній вік жертв')
plt.title('Середній вік жерта по місяцях')
plt.show()
```

```python
import numpy as np
```

```python
sns.lineplot(x="Month", y="Vict Age", data=df, hue = "Vict Sex", estimator=np.max) 
```

```python
sns.lineplot(data=df, x='Month', y='Vict Age', hue='Vict Sex', estimator=np.max)
plt.xlabel('Місяць')
plt.ylabel('Максимальний вік жертв')
plt.grid()
plt.show()
```

```python
sns.relplot(x="Month", y="Vict Age", data=df, kind='line', col="Vict Sex")
plt.show()
```

```python
sns.histplot(df['Vict Age'])
```

```python
sns.histplot(df['Vict Age'], kde=True)
```

```python
sns.histplot(df['Vict Age'], kde=True)
sns.set_theme(style="darkgrid")
plt.show()
```

```python
sns.histplot(df[df['Vict Sex'] == 'M']['Vict Age'], kde=True, label='M', color='blue')
sns.histplot(df[df['Vict Sex'] == 'F']['Vict Age'], kde=True, label='F', color='red')

plt.legend(title='Стать')
plt.xlabel('Вік жертв')
plt.ylabel('Кількість')
plt.title('Розподіл віку жертв за статтю')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.stripplot(data=df, x="Month", y="Vict Age", hue="Vict Sex", jitter=0.4, palette=sns.color_palette(["firebrick", "navy"]))

plt.legend(title='Стать', loc='upper right', bbox_to_anchor=(1.15, 1))
plt.xlabel('Місяць')
plt.ylabel('Вік жертв')
plt.title('Залежність місяця та віку жертв з розподілом за статтю')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.stripplot(data=df, x="Month", y="Vict Age", hue="Vict Sex", jitter=0.4, palette=sns.color_palette(["firebrick", "navy"]), dodge=True)

plt.legend(title='Стать', loc='upper right', bbox_to_anchor=(1.15, 1))
plt.xlabel('Місяць')
plt.ylabel('Вік жертв')
plt.title('Залежність місяця та віку жертв з розподілом за статтю')
plt.show()
```
