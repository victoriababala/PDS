```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 4 variant
x = np.arange(-10,40,0.1)
```

```python
y1 = 3.1875*x - 46.5625 
y2 = (5.8/16)*x + 83/16
y3 = x*(-103/73) + 2340/73
y4 = np.maximum(y3,y1)
```

```python
plt.plot(x,y1, ls='dotted',lw = 3)
plt.plot(x,y2, ls='--',lw = 3)
plt.plot(x,y3, ls = '-.',lw = 3)
plt.grid()
plt.xlim(14,19)
plt.ylim(7,13)
plt.title('Task 1')
plt.xlabel('X')
plt.ylabel('Y')
plt.text(17.55, 9, "10,2x1 -3,2x2 = 149",fontweight="bold", rotation=65, color="black")
plt.text(15.53, 8, "10,3x1 + 7,3x2 = 234",fontweight="bold", rotation=-40, color="black")
plt.text(15.53, 11, "-5,8x1 + 16x2 = 83",fontweight="bold", rotation=10, color="black")
plt.fill_between(x, y4,y2, where =(x > 15) & (x < 18.3), color = 'gray')
plt.show()
```
