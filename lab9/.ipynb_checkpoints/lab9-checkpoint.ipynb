```python
data = pd.read_csv("lab9b.csv")
italy_data = data[data['Entity'] == 'Italy']
italy_data.head()
```

```python
fig = plt.figure(figsize=(12, 9))
ax_1 = fig.add_subplot(2, 2, 1)
ax_2 = fig.add_subplot(2, 2, 2)
ax_3 = fig.add_subplot(2, 1, 2)

plt.style.use('seaborn-v0_8-pastel')
ax_1.bar(italy_data['Year'],italy_data['Births'],color='purple')
ax_2.bar(italy_data['Year'],italy_data['Deaths'])
fr_len = np.arange(len(italy_data))
ax_3.bar(fr_len-0.2, italy_data["Births"], width=0.4, color='purple')
ax_3.bar(fr_len+0.2, italy_data["Deaths"], width=0.4)
ax_3.set_xticks(ticks=fr_len, labels=italy_data['Year'])
ax_1.set_title('')
ax_2.set_title('')
ax_3.set_title('')
plt.subplots_adjust(wspace=0.2, hspace=0.5)
```

```python
fig = plt.figure(figsize=(12, 9))
ax_1 = fig.add_subplot(2, 2, 1)
ax_2 = fig.add_subplot(2, 2, 2)
ax_3 = fig.add_subplot(2, 1, 2)

plt.style.use('seaborn-v0_8-pastel')
ax_1.bar(italy_data['Year'],italy_data['Births'],color='purple')
ax_2.bar(italy_data['Year'],italy_data['Deaths'])
fr_len = np.arange(len(italy_data))
ax_3.bar(fr_len-0.2, italy_data["Births"], width=0.4, color='purple')
ax_3.bar(fr_len+0.2, italy_data["Deaths"], width=0.4, color='blue')
ax_3.set_xticks(ticks=fr_len, labels=italy_data['Year'])
ax_1.set_title('')
ax_2.set_title('')
ax_3.set_title('')
plt.subplots_adjust(wspace=0.2, hspace=0.5)
rects =ax_1.bar(italy_data['Year'], italy_data['Births'], color='indigo')
for rect in rects:
    height = rect.get_height()
    ax_1.text(rect.get_x() + rect.get_width()/2., 0.35*height,'%d' % round(int(height),-4),ha='center', va='bottom', color='white', rotation=90)

rects1 =ax_2.bar(italy_data['Year'], italy_data['Deaths'], color='blue')
for rect1 in rects1:
    height = rect1.get_height()
    ax_2.text(rect1.get_x() + rect1.get_width()/2., 0.35*height,'%d' % round(int(height),-4),ha='center', va='bottom', color='white', rotation=90)
```

```python
import plotly.express as px

fig = px.bar(data[data['Continent'] == 'Europe'], x='Entity', y = 'Births', color='Entity')
fig.show()
```

```python
fig = px.bar(data[data['Continent'] == 'Europe'], x='Entity', y = 'Births', color='Entity', animation_frame="Year", animation_group="Entity")
fig.show()
```

```python
fig = px.bar(data[data['Continent'] == 'Europe'], x='Entity', y = 'Births', color='Entity', animation_frame="Year", animation_group="Entity", range_y=[0,800000])
text=data[data['Continent']=='Europe']['Births'].apply(lambda x:f'{x/1000:.1f}')
fig.show()
```

```python
fig = px.sunburst(data, path=['Continent', 'Entity','Year'], values='Births')
fig.show()
```

```python
fig = px.sunburst(data, path=['Continent', 'Entity','Year'], values='Births', color_continuous_scale="BrBG",color='Births')
fig.update_traces(textinfo="label+percent parent")
fig.show()
```

```python
fig = px.treemap(data, path=['Continent','Entity','Year'], values = 'Births', color_continuous_scale='BrBG', color='Births')
fig.update_traces(textinfo='label+percent parent')
fig.show()
```

```python
fig = px.icicle(data, path=[px.Constant('world'), 'Continent', 'Entity', 'Year'], values='Births',                              
                  color_continuous_scale="BrBG",
                  color='Births')
fig.update_traces(textinfo="label+percent parent")
fig.show()
```

```python
fig = px.choropleth(data, locations="Code", color="Births", 
              hover_name="Entity",             
              color_continuous_scale=px.colors.sequential.Viridis, animation_frame="Year")
fig.show()
```

```python
fig = px.scatter_geo(data, locations="Code",  projection='orthographic', hover_name='Entity',  size="Births", color="Births", animation_frame = 'Year')
fig.show()
```
